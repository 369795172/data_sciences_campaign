# Linear Regression Analysis

Linear Regression Analysis Matrix Approach(Both Simple and Multiple)

y=b0 + b1x1 + b2x2 + ........ +bn*xn

```python
#Simple Linear regression (y=b0 + b1*x)
from numpy import *
points=genfromtxt("../../DataSets/pybk017master/data.csv",delimiter=",")
print(points)
```

```python
points=genfromtxt('../../DataSets/pybk017master/data.csv',delimiter=",")
points=delete(points,(0),axis=0)
y_actual=array(points[:,len(points[0])-1])
points[:,len(points[0])-1]=1
y_actual.resize(len(points),1)
beta=matrix(linalg.lstsq(points,y_actual)[0])
beta.resize(len(points[0]),1)
y_pred=dot(points,beta)
pro=(y_actual-y_pred)
error=sum(multiply(pro,pro))/len(points)
print("b0=",float(beta[len(beta)-1]),sep="")
for i in range(len(beta)-1):
    print("b",i+1,"=",float(beta[i]),sep="")
print("Error:",error)  
```
